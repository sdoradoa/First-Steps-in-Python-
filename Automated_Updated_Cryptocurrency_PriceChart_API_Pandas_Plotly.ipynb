```python
#Asking the cryptocurrency for the API, lower case needed.
name=input('Write a cryptocurrency lower case fullname:')
Name=name.capitalize()

from pycoingecko import CoinGeckoAPI #Import the library
import pandas as pd

cg=CoinGeckoAPI() #Create a client object
crypto_data= cg.get_coin_market_chart_by_id(id=name, vs_currency='usd', days=60) #Get the infotmation with the client
                                                                                 #and a function.
    
crypto_price_data=crypto_data['prices'] #Select the variable price among the data obtained
                                        #This gives a list with nested lists each one with time and price.
    
data=pd.DataFrame(crypto_price_data, columns=['TimeStamp','Price']) #Convert data to a pandas data frame
data['Date']= pd.to_datetime(data['TimeStamp'], unit='ms') #Use pandas method for time unit transformation
#data                               #->this prints the data frame its not needed thats why it is commented

import plotly.graph_objects as go #needed for the plot, a candlestick graph is selected and configured.

candlestick_data=data.groupby(data.Date.dt.date).agg({'Price':['min','max','first','last']})
fig = go.Figure(data=[go.Candlestick(x=candlestick_data.index, open=candlestick_data['Price']['first'], high=candlestick_data['Price']['max'],low=candlestick_data['Price']['min'], close=candlestick_data['Price']['last'])])
fig.update_layout(xaxis_rangeslider_visible=False, xaxis_title='Date', yaxis_title='Price (USD $)', title=Name+' Candlestick Chart Over Past 60 Days')

fig.show()
```
